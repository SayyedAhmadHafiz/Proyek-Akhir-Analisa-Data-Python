Nah, sekarang kita akan memasuki proses terakhir dalam data wrangling yaitu pembersihan atau cleaning data. Pada tahap ini, kita akan membersihkan berbagai masalah yang telah teridentifikasi dalam proses assessing data.

# 4.1 Membersihkan Data hours_df
Berdasarkan hasil proses assessing data, diketahui bahwa terdapat satu masalah yang dijumpai dalam hours_df, yaitu kesalahan type data. Pada tahap ini, kita akan membersihkan masalah tersebut.

```python
datetime_columns = ["dteday"]
 
for column in datetime_columns:
  hours_df[column] = pd.to_datetime(hours_df[column])
```

# - Visualisasi Data
# 1. Kapan waktu paling sibuk untuk penyewaan sepeda dalam sehari?

```python
# Menggunakan groupby untuk mengelompokkan berdasarkan 'hr' (jam)
penyewaan_per_jam = hours_df.groupby('hr')['cnt'].sum()

# Menyimpan data dalam DataFrame baru
penyewaan_per_jam_df = pd.DataFrame({'Jam': penyewaan_per_jam.index, 'Total Penyewaan': penyewaan_per_jam.values})

# Menggunakan idxmax() untuk menemukan jam dengan total penyewaan tertinggi
jam_tertinggi = penyewaan_per_jam_df['Total Penyewaan'].idxmax()

# Menampilkan hasil
print("Jam paling sibuk untuk penyewaan sepeda:", jam_tertinggi)

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(penyewaan_per_jam_df['Jam'], penyewaan_per_jam_df['Total Penyewaan'], color='lightblue')
plt.xlabel('Jam')
plt.ylabel('Total Penyewaan')
plt.title('Total Penyewaan Sepeda per Jam')

# Menyorot jam tertinggi dengan warna yang berbeda (warna pekat)
bars[jam_tertinggi].set_color('blue')

plt.grid(True)

# Set label-label pada sumbu x
plt.xticks(penyewaan_per_jam_df['Jam'])
plt.show()
```

# 2. Adakah tren peningkatan atau penurunan dalam jumlah total penyewaan sepeda?

```python
# Mengelompokkan data berdasarkan tanggal dan menghitung jumlah total penyewaan
daily_rentals = hours_df.groupby('dteday')['cnt'].sum()

# Mengonversi indeks (tanggal) menjadi format datetime
daily_rentals.index = pd.to_datetime(daily_rentals.index)

# Membuat plot tren peningkatan atau penurunan
plt.figure(figsize=(12, 6))
plt.plot(daily_rentals.index, daily_rentals)
plt.title('Tren Peningkatan atau Penurunan Penyewaan Sepeda')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)

# Menampilkan plot
plt.show()
```

# 3. pola penyewaan sepeda selama berjam-jam (hr) dan berbulan-bulan (mnth)

```python
# Analisis Jam ke Jam
hourly_rentals = hours_df.groupby('hr')['cnt'].sum()
plt.figure(figsize=(12, 6))
plt.plot(hourly_rentals.index, hourly_rentals.values)
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Total Penyewaan')
plt.title('Total Penyewaan Sepeda per Jam dalam Sehari')
plt.grid(True)
plt.show()

# Analisis Bulan ke Bulan
monthly_rentals = hours_df.groupby('mnth')['cnt'].sum()
plt.figure(figsize=(12, 6))
plt.bar(monthly_rentals.index, monthly_rentals.values)
plt.xlabel('Bulan dalam Setahun')
plt.ylabel('Total Penyewaan')
plt.title('Total Penyewaan Sepeda per Bulan dalam Setahun')
plt.grid(True)
plt.show()
```

# 4. Perbandingan Rata-rata Penyewaan Selama Hari Libur dan Hari Kerja

```python
# Kelompokkan data berdasarkan kolom "holiday"
holiday_data = hours_df[hours_df['holiday'] == 1]
non_holiday_data = hours_df[hours_df['holiday'] == 0]  # Data hari kerja (bukan hari libur)

# Hitung rata-rata penyewaan selama hari libur dan hari kerja   
average_rental_holiday = holiday_data['cnt'].mean()
average_rental_non_holiday = non_holiday_data['cnt'].mean()

# Visualisasikan perbandingan
plt.figure(figsize=(8, 4))
plt.bar(['Holiday', 'Non-Holiday'], [average_rental_holiday, average_rental_non_holiday])
plt.xlabel('Tipe Hari')
plt.ylabel('Rata-rata Penyewaan')
plt.title('Perbandingan Rata-rata Penyewaan Selama Hari Libur dan Hari Kerja')
plt.grid(True)
plt.show()
```
